```python
from selenium import webdriver
import time
import pandas as pd


url_str="https://www.tripadvisor.com/Attraction_Review-g294197-d324888-Reviews-or"
# url_num=str(i*5)
url_keyword="-Gyeongbokgung_Palace-Seoul.html"
driver = webdriver.Chrome("C:\\Users\\TJ\\Desktop\\chromedriver\\chromedriver.exe")

#페이지수늘려보기,외국어
for i in range(100):
    url_str2 = "https://www.tripadvisor.com/Attraction_Review-g294197-d324888-Reviews-"+ "or"+str(i*5)+ "-Gyeongbokgung_Palace-Seoul.html"
    driver.get(url_str2)
    time.sleep(5)
    if i == 0:
        driver.find_elements_by_css_selector(".continue")[i].click()
    reviews = driver.find_elements_by_css_selector(".location-review-review-list-parts-ExpandableReview__containerStyles--1G0AE")
    print([i])
    for k in range(len(reviews)):
        Temp_review = reviews[k].find_element_by_css_selector(".cPQsENeY").text
        review = Temp_review.replace("\n\n\n\n\n","                  ")
        print(review)
```
